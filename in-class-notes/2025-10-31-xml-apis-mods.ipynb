```python
MODS_collection_5 = Path('..','data','xml','2018_lcwa_MODS_5.xml')
```

```python
mods = etree.parse(MODS_collection_5)
metadata = mods.getroot()
```

```python
for item in metadata:
    print(item.tag)
```

```python
el_count = 0

for element in metadata.iter():
    el_count += 1
    print(element.tag)

print(el_count)
```

```python
ns = {
    'mods' : 'http://www.loc.gov/mods/v3',
    'ead3' : 'http://ead3.archivists.org/schema/'
}
```

Look through and find the identifiers for each of the 5 records: 

```python
for identifier in metadata.findall('.//mods:identifier', namespaces=ns):
    element = identifier
    print(element.tag, element.attrib, element.text)
```

```python
for identifier in metadata.findall('.//mods:identifier[@type="uri"]', namespaces=ns):
    print(identifier.tag, identifer.attrib, identifier.text)
```

Looking for `title` element and the text:

```python
for title in metadata.findall('.//mods:titleInfo/mods:title',namespaces=ns):
    print(title.tag, title.text, title.attrib)
```

```python
for mods in metadata.findall('.//mods:mods',ns):
    print('New Record:\n\t')
    for title in mods.findall('.//mods:title',ns):
        print(title.tag, title.text)
    print('\n')
```
